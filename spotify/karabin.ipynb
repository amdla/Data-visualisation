```python
# read file data.csv
import pandas as pd
data = pd.read_csv('data.csv')
text = data['text']
data = data.drop(columns=['text', 'Similar Song 1', 'Similar Song 2', 'Similar Song 3', 'Similar Artist 1', 'Similar Artist 2', 'Similar Artist 3', 'Similarity Score 1', 'Similarity Score 2', 'Similarity Score 3'])


data
```

```python
word_counts = {}
for song_text in text:
    words = song_text.lower().split()
    for word in words:
        word_counts[word] = word_counts.get(word, 0) + 1

# Sort the dictionary by value
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
# Create a DataFrame from the sorted word counts
word_counts_df = pd.DataFrame(sorted_word_counts, columns=['Word', 'Count'])

word_counts_df
```
