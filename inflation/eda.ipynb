```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

# Load the data
df = pd.read_csv('Economic Indicators And Inflation.csv')  # replace 'your_file.csv' with your actual file name

# Remove leading spaces from column names
df.columns = [col.strip() for col in df.columns]

# Filter data up to 2023
df_up_to_2023 = df[df['Year'] <= 2023]

# Filter data for 2023
df_2023 = df_up_to_2023[df_up_to_2023['Year'] == 2023]

# Sort countries by GDP in 2023
df_2023 = df_2023.sort_values(by='GDP (in billion USD)')

# Create 4 plots with 5 countries each
countries = df_2023['Country'].tolist()
gdp_values = df_2023['GDP (in billion USD)'].tolist()
```

```python
# Plot 1: Lowest GDP countries
plt.figure(figsize=(14, 6))
for country in countries[:4]:
    country_df = df_up_to_2023[df_up_to_2023['Country'] == country]
    plt.plot(country_df['Year'], country_df['GDP (in billion USD)'], label=country)
plt.xlabel('Year', fontsize=16, fontweight='bold')
plt.ylabel('GDP (in billion USD)', fontsize=16, fontweight='bold')
plt.title(f'GDP Over Time (Lowest GDP: {gdp_values[0]:.2f} - {gdp_values[3]:.2f} billion USD)', fontsize=16, fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), fontsize=14)
plt.grid(True, linestyle='-', alpha=0.5)  # Add grid
plt.tight_layout(rect=[0,0,0.8,1])  # Make space for the legend
plt.show()

# Plot 2: Low GDP countries
plt.figure(figsize=(14, 6))
for country in countries[4:9]:
    country_df = df_up_to_2023[df_up_to_2023['Country'] == country]
    plt.plot(country_df['Year'], country_df['GDP (in billion USD)'], label=country)
plt.xlabel('Year', fontsize=14, fontweight='bold')
plt.ylabel('GDP (in billion USD)', fontsize=14, fontweight='bold')
plt.title(f'GDP Over Time (Low GDP: {gdp_values[4]:.2f} - {gdp_values[8]:.2f} billion USD)', fontsize=16, fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), fontsize=14)
plt.grid(True, linestyle='-', alpha=0.5)  # Add grid
plt.tight_layout(rect=[0,0,0.8,1])  # Make space for the legend
plt.show()

# Plot 3: Medium GDP countries
plt.figure(figsize=(14, 6))
for country in countries[9:14]:
    country_df = df_up_to_2023[df_up_to_2023['Country'] == country]
    plt.plot(country_df['Year'], country_df['GDP (in billion USD)'], label=country)
plt.xlabel('Year', fontsize=14, fontweight='bold')
plt.ylabel('GDP (in billion USD)', fontsize=14, fontweight='bold')
plt.title(f'GDP Over Time (Medium GDP: {gdp_values[9]:.2f} - {gdp_values[13]:.2f} billion USD)', fontsize=16, fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), fontsize=14)
plt.grid(True, linestyle='-', alpha=0.5)  # Add grid
plt.tight_layout(rect=[0,0,0.8,1])  # Make space for the legend
plt.show()

# Plot 4: Highest GDP countries
plt.figure(figsize=(14, 6))
for country in countries[14:]:
    country_df = df_up_to_2023[df_up_to_2023['Country'] == country]
    plt.plot(country_df['Year'], country_df['GDP (in billion USD)'], label=country)
plt.xlabel('Year', fontsize=14, fontweight='bold')
plt.ylabel('GDP (in billion USD)', fontsize=14, fontweight='bold')
plt.title(f'GDP Over Time (Highest GDP: {gdp_values[14]:.2f} - {gdp_values[-1]:.2f} billion USD)', fontsize=16, fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), fontsize=14)
plt.grid(True, linestyle='-', alpha=0.5)  # Add grid
plt.tight_layout(rect=[0,0,0.8,1])  # Make space for the legend
plt.show()
```

```python
import matplotlib.cm as cm

colors = cm.tab20(np.linspace(0, 1, len(df_up_to_2023["Country"].unique())))

# Bar chart
plt.figure(figsize=(14, 6))
bottom = np.zeros(len(df_up_to_2023["Year"].unique()))
for i, country in enumerate(df_up_to_2023["Country"].unique()):
    country_df = df_up_to_2023[df_up_to_2023["Country"] == country]
    plt.bar(country_df["Year"], country_df["Inflation Rate (%)"], label=country, color=colors[i], bottom=bottom)
    bottom += np.array(country_df["Inflation Rate (%)"])
plt.xlabel("Year", fontsize=14, fontweight='bold')
plt.ylabel("Inflation Rate (%)", fontsize=14, fontweight='bold')
plt.title("Inflation Rate Over Time", fontsize=16, fontweight='bold')
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout(rect=[0,0,0.8,1])
plt.show()
```

```python
# Filter the data to only include the 2023 year
df_2023 = df_up_to_2023[df_up_to_2023['Year'] == 2023]

# Calculate the GDP Inflation Product for each country in 2023
df_2023['GDP_Inflation_Product'] = df_2023['GDP (in billion USD)'] * df_2023['Inflation Rate (%)']

# Sort the countries by the GDP Inflation Product in 2023
df_2023_sorted = df_2023.sort_values(by='GDP_Inflation_Product')

# Divide the countries into two groups based on the midpoint
mid_point = len(df_2023_sorted) // 2

# Select the countries around the midpoint
df_2023_group1 = df_2023_sorted.iloc[:mid_point]
df_2023_group2 = df_2023_sorted.iloc[mid_point:]
```

```python
# Create a scatter plot for the first group
plt.figure(figsize=(10, 6))
for country in df_2023_group1["Country"].unique():
    country_df = df_2023_group1[df_2023_group1["Country"] == country]
    plt.plot(country_df["GDP (in billion USD)"], country_df["Inflation Rate (%)"], label=country, marker='o')
plt.xlabel("GDP (in billion USD)")
plt.ylabel("Inflation Rate (%)")
plt.title("GDP vs Inflation Rate (Lower GDP*Inflation Rate)")
plt.legend()
plt.grid(True, linestyle='-', alpha=0.5)  # Add grid
plt.show()
```

```python
# Create a scatter plot for the second group
plt.figure(figsize=(10, 6))
for country in df_2023_group2["Country"].unique():
    country_df = df_2023_group2[df_2023_group2["Country"] == country]
    plt.plot(country_df["GDP (in billion USD)"], country_df["Inflation Rate (%)"], label=country, marker='o')
plt.xlabel("GDP (in billion USD)")
plt.ylabel("Inflation Rate (%)")
plt.title("GDP vs Inflation Rate (Higher GDP*Inflation Rate)")
plt.legend()
plt.grid(True, linestyle='-', alpha=0.5)  # Add grid
plt.show()
```
